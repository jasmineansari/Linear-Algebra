```python
def replace_missing_values(df, nv):
    for i, row in df.iterrows():
        missing_indices = np.where(pd.isna(row.values))[0]
        if len(missing_indices)>0:
            ki=[]
            for index in range(0,df.shape[1]):
                if index not in missing_indices:
                    ki.append(index)
            #x2 = -(nv[0] * row[0] + nv[2]*row[2])/nv[1]
            for missing_index in missing_indices:
                for null_vector in nv:
                    sum_product = 0
                    for j in ki:
                        sum_product +=null_vector[j]*row[j]
                    missing_value = -sum_product/null_vector[missing_index]
                    df.at[i,df.columns[missing_index]] = missing_value
    return df
```

```python
data = replace_missing_values(data, null_vectors)
```
